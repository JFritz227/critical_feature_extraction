```python
def buy_sell_points(csv, epsilon):
    try:
        close = pd.read_csv("data/sandp500/individual_stocks_5yr/" + csv)["Close"].values
            
        piece, lines, points = PLR(close, epsilon)
            
        buy_true = []
        period = math.floor(min(np.diff(points[1:]))/2)
            
        for the_points in points[1:]:
            if piece[the_points] - piece[the_points + period] > 0:
                buy_true.append(-1)
            if piece[the_points] - piece[the_points + period] < 0:
                buy_true.append(1)
                    
        lists = up_down_trend(close, 20)
        close_list = []
        x_list = []
            
        for list_number in range(len(lists)):
            x_axis = points[1:]
            x_list.append(x_axis)
            close_list.append(close[x_axis])
                
        technicals = pd.read_csv("data/sandp500/individual_stocks_5yr_TECHNICALS/" + csv)
            
        for x_days in range(len(x_list[0])):
            for days in technicals['Unnamed: 0']:
                if x_list[0][x_days] == days:
                    technicals.loc[technicals['Unnamed: 0'] == x_list[0][x_days], 'target'] = buy_true[x_days]
        technicals.loc[technicals['target'].isnull(), 'target'] = 0
            
        dummies = pd.get_dummies(technicals['target'])
        technicals['buy'] = dummies[1.0]
        technicals['sell'] = dummies[-1.0]
        technicals = technicals.drop('target', 1)
        return technicals
    
    except:
        if epsilon < 100:
            epsilon += 5
            buy_sell_points(csv, epsilon)
        elif epsilon >= 100:
            print(csv)
```

```python
individuals = os.listdir(path = "data/sandp500/individual_stocks_5yr/")
for stock in individuals:
    try:
        new_df = buy_sell_points(stock, 5)
        new_df.to_csv("data/sandp500/individual_stocks_5yr_buy_or_sell/" + stock)
    except:
        pass
```
